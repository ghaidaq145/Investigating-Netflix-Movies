```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
#filtering the 1990s films
net_1990s=netflix_df[np.logical_and(netflix_df['release_year']<2000,netflix_df['release_year']>=1990)]
```

```python
#visualizing the data to see the mosr frequent duration apparently its between 90 and 100 lets test these values one of them must have above 5 iterations 
plt.hist(net_1990s['duration'], bins=200)
plt.xlabel('duration_in_minutes')
plt.ylabel('frequency')
plt.title('the most frequent films in the 1990s')
plt.yticks([0,5,10,15])
plt.show()
duration= 94
```

```python
#testing the values between 90 and 100 it appears that 94 have 7 iterations and thats reasonable 
freq = 0
for lab, row in net_1990s.iterrows():
    if net_1990s.loc[lab, 'duration'] == 94:
        freq = freq + 1
print(freq)
print("the most frequent 90s film duration in netflix is:", duration)
```

```python
#flitering the Action films
net_short=net_1990s[net_1990s['genre']=='Action']
#calculating the action films which have a short duration (<90)
short_movie_count=0
for lab,row in net_short.iterrows():
    if net_short.loc[lab,'duration']<90:
        short_movie_count=short_movie_count+1
print("the number of short action films in netflix is:",short_movie_count)
```
